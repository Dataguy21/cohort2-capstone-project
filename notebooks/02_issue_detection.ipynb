```python
"""
Capstone 3 - Clinical Insights Assistant
Notebook: 02_issue_detection.ipynb
------------------------------------
Detects and visualizes:
- Non-compliance patterns
- Adverse events
- Outcome anomalies
- Summary of overall trial issues
"""

import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Import functions from src
import sys
sys.path.append(os.path.abspath(os.path.join(os.getcwd(), "../src")))
from issue_detection import (
    detect_non_compliance,
    detect_adverse_events,
    detect_outcome_anomalies,
    summarize_issues
)

# ------------------- CONFIG -------------------
pd.set_option("display.max_columns", None)
sns.set(style="whitegrid")

# ‚úÖ Universal data path
project_root = os.getcwd()
if "notebooks" in project_root:
    DATA_PATH = os.path.abspath(os.path.join(project_root, "../data/clinical_trial_data.csv"))
else:
    DATA_PATH = os.path.abspath(os.path.join(project_root, "data/clinical_trial_data.csv"))

print("Using dataset:", DATA_PATH)

# ------------------- LOAD DATA -------------------
df = pd.read_csv(DATA_PATH, parse_dates=["visit_date"])
df.columns = [c.strip().lower().replace(" ", "_") for c in df.columns]
print("‚úÖ Dataset loaded successfully!")
print(f"Shape: {df.shape}")

# ------------------- 1. Non-Compliance Detection -------------------
non_compliant_df = detect_non_compliance(df, threshold=80)
print(f"üîé Non-Compliant Records: {len(non_compliant_df)}")

if not non_compliant_df.empty:
    display(non_compliant_df.head(10))

    plt.figure(figsize=(8, 5))
    sns.histplot(non_compliant_df["compliance_rate"], bins=15, kde=True, color="tomato")
    plt.title("Distribution of Non-Compliance Rates", fontsize=14)
    plt.xlabel("Compliance Rate (%)")
    plt.ylabel("Count")
    plt.show()
else:
    print("‚úÖ No non-compliance detected in dataset.")

# ------------------- 2. Adverse Event Detection -------------------
adverse_df = detect_adverse_events(df)
print(f"‚ö†Ô∏è Adverse Event Records: {len(adverse_df)}")

if not adverse_df.empty:
    display(adverse_df.head(10))

    plt.figure(figsize=(6, 4))
    event_trend = adverse_df.groupby("visit_date")["adverse_event"].sum().reset_index()
    sns.lineplot(x="visit_date", y="adverse_event", data=event_trend, color="red")
    plt.title("Adverse Events Over Time", fontsize=14)
    plt.xlabel("Visit Date")
    plt.ylabel("Event Count")
    plt.xticks(rotation=45)
    plt.show()
else:
    print("‚úÖ No adverse events recorded in dataset.")

# ------------------- 3. Outcome Anomaly Detection -------------------
anomaly_df = detect_outcome_anomalies(df)
print(f"üìä Outcome Anomalies Detected: {len(anomaly_df)}")

if not anomaly_df.empty:
    display(anomaly_df.head(10))

    plt.figure(figsize=(8, 5))
    sns.scatterplot(
        x="dosage_mg", y="outcome_score", hue="anomaly_type",
        data=anomaly_df, palette={"Low Outcome": "red", "High Outcome": "green"}
    )
    plt.title("Outcome Anomalies by Dosage", fontsize=14)
    plt.xlabel("Dosage (mg)")
    plt.ylabel("Outcome Score")
    plt.show()
else:
    print("‚úÖ No significant outcome anomalies detected.")

# ------------------- 4. Summary of Issues -------------------
summary = summarize_issues(df)
print("\nüìã Issue Summary:")
for k, v in summary.items():
    print(f" - {k.replace('_', ' ').title()}: {v}")

# ------------------- 5. Visualization: Issue Counts -------------------
plt.figure(figsize=(6, 4))
sns.barplot(
    x=["Non-Compliance", "Adverse Events", "Outcome Anomalies"],
    y=[summary["non_compliance_count"], summary["adverse_event_count"], summary["outcome_anomaly_count"]],
    palette="pastel"
)
plt.title("Summary of Detected Issues", fontsize=14)
plt.ylabel("Record Count")
plt.show()

print("\n‚úÖ Issue Detection and Visualization completed successfully!")
```
